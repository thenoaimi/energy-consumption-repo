```python
import pyreadr
import pandas as pd
import numpy as np
```

```python
# reading from rda file and converting the dataset to a pandas data frame

data = pyreadr.read_r('./Data/data/gefcom.rda')
df = data['gefcom']
```

```python
# Getting the data for Connecticut (CT)
dfCT = df.loc[df['zone'] == 'CT']
# Taking a sample of five years for the regression model to avoid over fitting
dfCT = dfCT.loc[(dfCT['year'] >= 2004.0) & (dfCT['year'] <= 2016.0)]
#dfCT = dfCT.loc[(dfCT['year'] <= 2016.0)]
dfCT = dfCT.drop(['holiday_name'], axis=1)
CT =dfCT.set_index('ts')
CT.head()
```

```python
# Checking for missing entries within the dataset
CT.groupby(CT['year']).count()
```

In a year there should be 8760 hours and 8784 hours in a leap year. This means that there are two missing hours in each year.

```python
search = CT.loc[(CT['year'] == 2015.0) & (CT['month'] == 'Nov')]
search
```

We can see the the dataset is missing the value, "2015-11-01 01:00:00". We need to insure the timeseries is continuous and not missing any values.

```python
# Setting index frequency for the time series to be in 1 hour intervals
s = CT.asfreq(freq='1h')
# Reseting the index
s.reset_index(inplace=True)
s['ts'] = s['ts'].dt.floor('H')
s.groupby(s['ts'].dt.year).count()
```

We can see that the time is now fixed with the correct number of hours, however we have added null values in the data which need to be imputed

```python
# Observing the missing values
s[s.isna().any(axis=1)]
```

```python
# Setting the zone to CT as this value will not change for the entirety of the dataset
s['zone'] = s['zone'].fillna(value='CT')
# Extracting the date
s['date'] = s['ts'].dt.date
# Extracting the hour
s['hour'] = s['ts'].dt.hour
# Extracting the month
s['month'] = s['ts'].dt.month
# Extracting the year
s['year'] = s['ts'].dt.year
# Extracting the day of year
s['day_of_year'] = s['ts'].dt.dayofyear
# Getting the day of week
weekdays = {0:'Mon', 1:'Tue', 2:'Wed', 3: 'Thu', 4: 'Fri', 5:'Sat', 6:'Sun'}
s['day_of_week'] = s['ts'].dt.dayofweek.map(weekdays)
# Creating a boolean value for weekend
s['weekend'] = s.apply(lambda x: True if \
                                         ((x['day_of_week'] in ['Sat', 'Sun']))
                                         else False, axis = 1)
```

```python
#Adding holidays column to indicate whether a day was a holiday as per the US calendar or not
from pandas.tseries.holiday import USFederalHolidayCalendar as calendar
cal = calendar()
holidays = cal.holidays(start = s['date'].min(), end = s['date'].max())
s['holiday'] = s['date'].isin(holidays)

#adding in another column which indicates 'non-working' days which include weekends and holidays
s['non_working'] = s.apply(lambda x: 'non-working' if \
                                         ((x['holiday'] == 1) or (x['day_of_week'] in ['Sat', 'Sun']))
                                         else 'working', axis = 1)



s.head()
```

```python
#Filling in the missing values by the average values from the same hour, weekday and month taken from rest of the data from that
# year
for date in s[s.isna().any(axis=1)].ts:
    s[s.ts == date] = s[s.ts== date].fillna\
                                                   (s[(s.ts.dt.year == date.year) &\
                                                               (s.ts.dt.month == date.month) & \
                                                               (s.ts.dt.weekday == date.dayofweek) &\
                                                               (s.ts.dt.hour == date.hour) \
                                                               ].mean())
```
